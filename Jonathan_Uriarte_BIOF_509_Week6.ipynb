We'll load the Pima Indians diabetes dataset. This is a binary classification task for predicting 5-year diabetes onset. The features are:

   1. Number of times pregnant
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
   3. Diastolic blood pressure (mm Hg)
   4. Triceps skin fold thickness (mm)
   5. 2-Hour serum insulin (mu U/ml)
   6. Body mass index (weight in kg/(height in m)^2)
   7. Diabetes pedigree function
   8. Age (years)

```python
dataset = loadtxt('pima-indians-diabetes.csv', delimiter=',')
import numpy
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**1. Define a Keras sequential model with two hidden layers, each with 64 neurons, and a single output neuron. Choose appropriate activation functions for each layer.**
- Hint: remember that the input shape must match the number of features we have.
- Hint: the activation function for the final (output) layer will depend on the type of task the network is doing.
- Johnny Note. Here is the website documentation to how to make this work [Keras Documentation](https://keras.io/guides/sequential_model/)
- Johnny Note (2): Another set of documentation [Tensor Flow](https://www.tensorflow.org/api_docs/python/tf/keras/Sequential)
- Johnny Note (3): Found a good explination websites [First Deep Learning Project](https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/)

```python
# YOUR CODE HERE (1)

# set up the sequential model
model = Sequential()

# set up the first layer with 64 neurons, also uses the 8 dimensions (columns) we'll be using from our dataset.
## we're using the 'relu' activation by default
model.add(Dense(64, input_dim=8, activation='relu'))

# This sets up the second layer with 64 neurons. No need for input_dim here.
## we're using the 'relu' activation by default
model.add(Dense(64, activation='relu'))

# This is the single output neuron
## This uses the 'sigmoid' activation function.
model.add(Dense(1, activation='sigmoid'))
```

**2. Compile the model with binary cross-entropy loss, accuracy for metrics, and a Stochastic Gradient Descent optimizer.**
- Hint: see the documentation for [compile](https://www.tensorflow.org/api_docs/python/tf/keras/Model#compile).

```python
# YOUR CODE HERE (2)
## sgd = stochastic gradient descent
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

**3. Fit the model to the training data, using 16 epochs and a batch size of 8.**
- Hint: see the documentation for [fit](https://www.tensorflow.org/api_docs/python/tf/keras/Model#fit).

```python
# YOUR CODE HERE (3)
model.fit(X_train, y_train, epochs=16, batch_size=8)
```

**4. Print the percent accuracy on the test data**

```python
# YOUR CODE HERE (4)
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```
